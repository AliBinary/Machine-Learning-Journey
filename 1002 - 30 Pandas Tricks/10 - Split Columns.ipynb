## Make Dataframe

```python
df = pd.DataFrame({'Name':['Ali Ghanbari', 'Zahra Ghalenave'], 'Location':['Tehran, Iran', 'Melbourn, Australia']})
df
```

```python
df[['Firstname', 'Lastname']] = df.Name.str.split(' ', expand=True)
df
```

## Alternative Solution

```python
df['Firstname'], df['Lastname'] = zip(*df.Name.str.split(' '))
df
```
