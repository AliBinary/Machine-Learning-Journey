```python
my_dict = {
            'Name': ['a','b','c','d'],
            'Score': [100,200,300,0.005]
          }

df = pd.DataFrame(my_dict)
df
```

```python
df.to_csv('./Output/my_file.csv')
```

```python
# Use Tab to Seperate data
df.to_csv('./Output/my_file2.csv',sep='\t')
```

```python
# rounded to two decimals
df.to_csv('./Output/my_file3.csv',float_format='%.2f')
```

```python
df.to_csv('./Output/my_file4.csv',header=False)
```
