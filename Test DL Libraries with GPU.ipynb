```python
# test_ds_env.py

import sys

print("="*50)
print("Python version:", sys.version)
print("="*50)

# ===== TensorFlow =====
try:
    import tensorflow as tf
    print("\n===== TensorFlow =====")
    print("Version:", tf.__version__)
    
    gpus_tf = tf.config.list_physical_devices('GPU')
    if gpus_tf:
        print("TensorFlow: GPUs detected!")
        for i, gpu in enumerate(gpus_tf):
            print(f"  GPU {i}: {gpu}")
    else:
        print("TensorFlow: No GPU detected (CPU mode)")
except ImportError:
    print("\nTensorFlow is not installed!")
except Exception as e:
    print("\nTensorFlow error:", e)

print("="*50)

# ===== PyTorch =====
try:
    import torch
    print("\n===== PyTorch =====")
    print("Version:", torch.__version__)
    
    if torch.cuda.is_available():
        print("PyTorch: CUDA is available âœ…")
        print("GPU name:", torch.cuda.get_device_name(0))
        print("Number of GPUs:", torch.cuda.device_count())
        print("CUDA version used by PyTorch:", torch.version.cuda)
    else:
        print("PyTorch: No GPU detected (CPU mode)")
except ImportError:
    print("\nPyTorch is not installed!")
except Exception as e:
    print("\nPyTorch error:", e)

print("="*50)
print("Environment check finished!")
```
