# AutoML

Machine learning is an application of artificial intelligence (AI) that provides systems the ability to automatically learn and improve from experience without being explicitly programmed.

In machine learning, **supervised** and **unsupervised learning** are the two major techniques.  
Supervised machine learning algorithms can apply what has been learned in the past to new data using labeled examples to predict future events, whereas unsupervised machine learning algorithms are used when the information used to train is neither classified nor labeled.



When applying machine learning models, we usually perform:

- Data preprocessing  
- Feature engineering  
- Feature extraction  
- Feature selection  

After this, we select the best algorithm and tune parameters to obtain the best results.  
To reduce this burden, **AutoML (Automatic Machine Learning)** tools are introduced.

The main benefit of these tools is reducing the burden of data cleaning and preprocessing.  
There are many AutoML packages available in Python.

> **Note:**  
> This kernel organizes regression, classification and clustering on one platform with only a few lines of code using the **PyCaret** package.

## About PyCaret



PyCaret is an **open-source, customizable, low-code AutoML library** in Python that aims to reduce the cycle time from hypothesis to insights.

It makes data scientists more productive by significantly reducing the amount of code required for experiments. PyCaret is essentially a Python wrapper around several machine learning libraries and frameworks such as:

- scikit-learn  
- XGBoost  
- Microsoft LightGBM  
- spaCy  

Through PyCaret, the time spent on coding experiments is drastically reduced.

### Model Saving & Prediction (General Guide)

- Trained models can be saved for **later use or deployment**
- Saved models include preprocessing, feature engineering, and the trained estimator
- Loaded models behave exactly like the original trained models

#### Common functions
- `save_model()`  persist a trained model to disk  
- `load_model()`  reload a previously saved model  
- `predict_model()`  generate predictions on new or unseen data

```python
save_model(dt, 'MyRegressor')
```

```python
loaded_model = load_model('MyRegressor')
```

```python
predict_model(loaded_model, data=X_test)
```

### PyCaret `tune_model` (General Guide)

- `models()` lists all available models in the current experiment
- `create_model()` trains a model using default hyperparameters
- `tune_model()` optimizes model hyperparameters using cross-validation
- Tuning aims to improve performance based on a selected metric

```python
models()
```

```python
# train a xgboost (Extreme Gradient Boosting) model with default parameters
xgboost = create_model('xgboost')
# tune hyperparameters of xgboost (Extreme Gradient Boosting) model
tuned_xgboost = tune_model(xgboost)
```

### PyCaret AutoML (General Guide)

- `setup()` initializes the experiment (data, target, preprocessing, CV)
- Models must be **trained first** using `create_model()` or `compare_models()`
- `automl()` **does NOT train new models**
- `automl()` selects the best model from **already trained models**
- Model selection is based on the chosen evaluation metric

```python
setup(df, target='medv')
compare_models()      # train all models
best = automl()       # select best trained model
```

```python
best
```

### PyCaret: Model Creation & Blending (Conceptual Overview)

**1. Model Creation**
- `create_model()` trains a model on your dataset.  
- Each model can be different (linear, tree-based, etc.) depending on the problem.  
- The key is to get multiple candidate models for experimentation.

**2. Model Blending**
- `blend_models()` combines predictions from multiple trained models into a single ensemble.  
- Concept: "wisdom of the crowd"  ensemble predictions are usually more robust.  
- Reduces overfitting and leverages the strengths of diverse models.

**3. Practical Tips**
- Start with a few diverse models for experimentation.  
- Use blending to improve overall performance instead of relying on a single model.  
- Works for both regression and classification tasks.


```python
rd = create_model('ridge');      
lgm  = create_model('lightgbm');            

# blending 3 models
blend = blend_models(estimator_list=[lgm, lr, rd])
```

```python
type(blend)
```

```python
type(lr)
```

### PyCaret: Optimizing Thresholds (Conceptual Overview)

**1. Purpose of Threshold Optimization**
- Classification models output probabilities for each class.  
- By default, a threshold of 0.5 is used to assign classes.  
- `optimize_threshold()` finds the threshold that maximizes a chosen metric (e.g., Recall, F1, Precision).

**2. Concept**
- Adjusting the threshold can improve performance for the metric you care about.  
- Useful especially for imbalanced datasets or when certain errors are more costly than others.  

**3. Practical Tips**
- Choose the metric (`optimize`) based on your goal.  
  - Example: use `Recall` if missing positive cases is costly.  
- Threshold optimization is lighter if you first experiment with simple models.  
- Visualizing the metric across different thresholds helps understand model behavior.


```python
optimize_threshold(lr, optimize='Recall')
```

### Model Evaluation

```python
evaluate_model(lr)
```

```python
plot_model(estimator = blend, plot = 'confusion_matrix')
```

```python
pred = predict_model(blend, data = X_test)
pred
```

```python
output = pd.DataFrame({'Survived': pred.prediction_label})
output.to_csv('submission.csv')
```
