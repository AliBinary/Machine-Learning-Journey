## ydata-profiling: Comprehensive Automated EDA

---



---

ydata-profiling generates standardized and rich exploratory data analysis reports. 
It goes beyond simple summaries by providing:
1. **Type Inference:** Automatically detects if a column is categorical, numerical, or text.
2. **Quantile & Descriptive Stats:** Mean, mode, median, skewness, and kurtosis.
3. **Correlations:** Highlighting highly correlated variables (Pearson, Spearman, Kendall).
4. **Missing Values:** Matrix, bar, and heatmap of nullity.

```python
import pandas as pd
from ydata_profiling import ProfileReport
```

###  Loading the Data

```python
df = pd.read_csv('./Data/Boston.csv')

# Basic info
print(f"Dataset Shape: {df.shape}")
df.head()
```

###  Create Profile Report

```python
profile = ProfileReport(df, title="Titanic Dataset Profiling Report", explorative=True)
```

###  Display Report in HTML

```python
# Save the profile report as HTML
profile.to_file("./Output/YDdata-Profiling_report.html")

# Open the saved HTML file in the default web browser
import webbrowser
import os

file_path = os.path.abspath("./Output/YDdata-Profiling_report.html")
webbrowser.open(f"file://{file_path}")
```
