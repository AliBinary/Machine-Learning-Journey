
## Sweetviz  Automated Exploratory Data Analysis (EDA)

Sweetviz is a Python library that helps you **quickly understand a dataset** by automatically generating
a rich **HTML EDA report** with minimal code.

It is especially useful at the **very beginning of a machine learning project**.
    





## What Sweetviz Does

Sweetviz automatically analyzes:
- Numerical feature distributions
- Categorical feature frequencies
- Missing values
- Feature correlations
- Feature behavior **relative to the target variable**

The result is a **single interactive HTML file** that you can:
- Open in a browser
- Share with teammates
- Archive as project documentation

##  Load Dataset

```python
import pandas as pd
import sweetviz as sv

df = pd.read_csv("./Data/Titanic-Dataset.csv")
df.head()
```


## Generate Sweetviz Report

With **one line of code**, Sweetviz analyzes the entire dataset.


```python
report = sv.analyze(df, target_feat="Survived")
report.show_html("./Output/SweetViz_report.html")
```


### Output

- An interactive HTML file is created
- Open it in your browser to explore:
  - Feature distributions
  - Missing data
  - Target impact

Sweetviz is **read-only EDA** (no manual plotting).
