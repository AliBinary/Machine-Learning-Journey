```python
fa_month = {"فروردین" : 1 , 
"اردیبهشت" : 2 , 
"خرداد" : 3 , 
"تیر" :  4, 
"مرداد" :  5, 
"شهریور" :  6, 
"مهر" :  7, 
"آبان" :  8, 
"آذر" :  9, 
"دی" : 10 , 
"بهمن" : 11 , 
"اسفند" :  12, 
}
```

```python
def jalali_to_m (date):
    year , day , month = date.split('-')
    changed_date = JalaliDate(int(year) ,fa_month[month]  , int(day)).to_gregorian()
    return changed_date
    
```

```python
df_numbers["MDate"] = df_numbers["Date"].apply(jalali_to_m)
```
